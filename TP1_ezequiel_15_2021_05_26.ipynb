# **2. Hallar π por dos caminos:**

## **a. Cálculo de π con método de Newton-Raphson usando seno(x)**

2.a.jpg




```python
def funcion_newton_raphson(x):
  return x - ( np.sin(x)/np.cos(x) )


def calcular_pi_newton_raphson(semilla):

  p_i = semilla

#DUDA criterio de corte -> "hasta alcanzar el limite de la herramienta utilizada" ¿?
  while( abs(np.sin(p_i)) > 10**(-15)):
    p_i = p_i - ( np.sin(p_i) / np.cos(p_i))

  return p_i
```

## **b. Cálculo de π con Serie de Leibniz**

2.b.jpg

A partir de la serie de Leibniz, definida como:

Serie Leibniz.jpg

Con un numero finito '***n***' de iteraciones, podemos hallar una aproximación de π multiplicando el resultado obtenido por 4.

```python
def calcular_pi_leibniz(n):

  pi = 0

  for i in range(0, n):
    pi += ((-1)**i ) / (2*i + 1)

  pi = 4 * pi
  return pi
```

## **c. Obtención de aproximaciones de π con Newton-Raphson y Leibniz: precisión simple (32 bits)**



2.c.jpg

```python
a = np.single(calcular_pi_con_leibniz(10))
b = np.single(calcular_pi_con_leibniz(100))
c = np.single(calcular_pi_con_leibniz(1000))
d = np.single(calcular_pi_con_leibniz(10000))
e = np.single(calcular_pi_con_leibniz(100000))

semilla = 3.0
f = np.single(calcular_pi_newton_raphson(semilla))

print("Aproximaciones por Serie de Leibniz:\n")
j = 1
for i in [a,b,c,d,e]:
  print("n =", 10**j,"->  π =", i)
  j += 1

print("\n\nAproximación por método de Newton-Raphson:\n")
print("Semilla =", semilla, "->  π =",f)
```
